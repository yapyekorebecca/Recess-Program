```python
# working on machine leraning problem with haert diseaes datasets
import pandas as pd
# load the dataset
data= pd.read_csv('heart.csv')
# Display the first few rows of the dataset
print(data)
```

```python
# data processing and data cleaning 
# check for missing values
missing_values= data.isnull().sum()
print(missing_values)
```

```python
# checking for duplicate values
print(data.duplicated().sum())

```

```python
data.drop_duplicates(inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the histogram
plt.figure(figsize=(10, 6))
# Use seaborn to plot histograms for different genders
sns.histplot(data=data, x='age', hue='sex', multiple='stack', kde=True)

# Customizing the plot
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'data' is your DataFrame and it has 'age', 'sex', and 'chol' columns

# Create scatter plot
plt.figure(figsize=(10, 6))

for gender in data['sex'].unique():
    subset = data[data['sex'] == gender]
    plt.scatter(subset['age'], subset['chol'], label=f'Sex: {gender}', alpha=0.6)

# Customize the plot
plt.title('Age vs. Cholesterol Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend(title='Gender', labels=['Female', 'Male'])  # Assuming 0 is Female and 1 is Male
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Select the columns you want to visualize
columns_to_plot = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Melt the DataFrame to long format
df_melted = data.melt(value_vars=columns_to_plot, var_name='Variable', value_name='Value')

# Plot the box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_melted, x='Variable', y='Value')

# Customizing the plot
plt.title('Distribution of Selected Variables')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.grid(True)

# Show the plot
plt.show()
```
